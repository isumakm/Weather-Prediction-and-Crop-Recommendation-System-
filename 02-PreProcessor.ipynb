```python
# Imports
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Define column types
categorical_cols = ['crop']
numeric_cols = [
    'temperature', 'rainfall', 'sunshine_hours', 'ph',
    'organic_carbon', 'cec', 'awc', 'bulk_density', 'texture_code'
]
```

```python
# Build preprocessor
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

display("Preprocessor pipeline created with numeric cols:", numeric_cols)
```

```python
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

print("and categorical cols:", categorical_cols)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ],
    remainder='drop'
)

print(preprocessor)
```
