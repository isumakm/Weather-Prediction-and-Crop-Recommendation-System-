```python
import pandas as pd
import json

df = pd.read_csv("../outputs/soil_clusters_top_3_models.csv")

FEATURES = ["taw", "organic_carbon", "cec", "ph", "sand_pct"]

CLUSTER_MODELS = {
    "kmeans": "cluster_kmeans",
    "agglomerative": "cluster_agg",
    "gmm": "cluster_gmm"
}
```

### **1. Export Cluster Map Data**

```python
for model, col in CLUSTER_MODELS.items():
    csv_path = f"../outputs/web_cluster_points_{model}.csv"
    df[["lat", "lon", col]].rename(columns={col: "cluster"}).to_csv(
        csv_path, index=False
    )
    print(f"Saved: {csv_path}")
```

### **2. Export Cluster Feature Means**

```python
cluster_means = {}

for model, col in CLUSTER_MODELS.items():
    cluster_means[model] = (
        df.groupby(col)[FEATURES]
        .mean()
        .round(3)
        .to_dict()
    )

with open("../outputs/web_cluster_means.json", "w") as f:
    json.dump(cluster_means, f, indent=2)
```
